```python
# p1307
def reverse_number(n):
    if n == '-0':
        return '-0'
    else:
        if not n or (len(n)==1 and n[0]=='-'):
            return n
        # 移除末尾的零
        while len(n)>1 and n[-1] == '0':
            n = n[:-1]
        # 如果是负数，则移除负号，并在最后添加回来
        is_negative = n[0] == '-'
        if is_negative:
            n = n[1:]
        r = n[::-1]
        if is_negative:
            r = '-' + r
        return r
n = input()
print(reverse_number(n))
```

```python
# p1720
def f(n):
    """返回斐波那契数列的第n位数字"""
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

n = int(input())
ans = f(n)
print('%.2f'%ans)
```

```python
# p5725
def print_matrices(n):
    if n > 9 or n < 1:
        print()
        return
    # 打印普通矩阵
    for i in range(n):
        for j in range(n):
            print(f'{i * n + j + 1:02d}', end='')
        print()
    print()
    # 打印右对齐的直角三角形矩阵
    counter = 1
    for i in range(n):
        print('  ' * (n - i - 1), end='')
        for j in range(i + 1):
            print(f'{counter:02d}', end='')
            counter += 1
        print()


n = int(input())
print_matrices(n)
```

```python
# p1089
def calculate_year_end_balance(budgets):
    allow = 300
    savings = 0
    balance = 0
    for month, budget in enumerate(budgets, start=1):
        balance += allow
        if balance < budget:
            return -month
        balance -= budget
        if balance >= 100:
            to_save = (balance // 100) * 100
            savings += to_save
            balance -= to_save

    savings_with_interest = int(savings * 1.2)
    year_end_balance = balance + savings_with_interest
    return year_end_balance

budgets = []
for _ in range(12):
    budget = int(input())
    if budget < 0 or budget >= 350:
        raise ValueError("预算必须是非负整数且小于350")
    budgets.append(budget)

result = calculate_year_end_balance(budgets)
print(result)
```

```python
# p5732
def triangle(n):
    l=[]
    for i in range(n):
        if i==0:
            l.append([1])
        elif i==1:
            l.append([1,1])
        else:
            y = []
            for j  in range(i+1):
                if j==0 or j==i:
                    y.append(1)
                else:
                    y.append(l[i-1][j]+l[i-1][j-1])
            l.append(y)
    return l
n = int(input())
x=triangle(n)
for i in range(len(x)):#逐行打印结果
    print(' '.join(map(str,x[i])))
```

```python
# p2615
n = int(input())
a = [[0 for _ in range(n)] for i in range(n)]
a[0][n//2] = 1
col,row = 0,n//2
for k in range(2,n**2 + 1):
    if col == 0 and row != n-1:
        row += 1
        col = n-1
        a[col][row] = k
    elif col != 0 and row == n-1:
        row = 0
        col -= 1
        a[col][row] = k
    elif col == 0 and row == n-1:
        col += 1
        a[col][row] = k
    elif col != 0 and row != n-1:
        if a[col-1][row+1] == 0:
            row +=1
            col -= 1
            a[col][row] = k
        else:
            col += 1
            a[col][row] = k
for i in range(n):
    for j in range(n):
        print(a[i][j],end=" ")
    print()
```

```python
# p2911
from collections import Counter
a, b, c = map(int, input().split())
li = []
for i in range(1, a+1):
    for j in range(1, b+1):
        for k in range(1, c+1):
            li.append(i + j + k)
ans = Counter(li)
max_count = max(ans.values())
most_frequent_sums = [sum_ for sum_, count in ans.items() if count == max_count]
print(min(most_frequent_sums))
```

```python
# p1161
import math
n = int(input())
lights = {}
for _ in range(n):
    a, t = map(float, input().split())
    t = int(t)
    for j in range(1,t+1):
        light_id = math.floor(j*a)
        if light_id in lights:
            lights[light_id] = not lights[light_id]
        else:
            lights[light_id] = True
for light_id,is_on in lights.items():
    if is_on:
        print(light_id)
        break
```

```python
n = int(input())
grids = [[0] * n for _ in range(n)]
t, l, r, b = 0, 0, n - 1, n - 1
num = 1
while True:
    for i in range(l, r + 1):  # 从左到右
        grids[t][i] = num
        num += 1
    t += 1
    if t > b: break
    for i in range(t, b + 1):  # 从上到下
        grids[i][r] = num
        num += 1
    r -= 1
    if r < l: break
    for i in range(r, l - 1, -1):  # 从右到左
        grids[b][i] = num
        num += 1
    b -= 1
    if b < t: break
    for i in range(b, t - 1, -1):  # 从下到上
        grids[i][l] = num
        num += 1
    l += 1
    if r < l: break
for i in range(n):
    print("".join(f"{x:3}" for x in grids[i]))
```

```python
# p1205
import copy  # 为了后续的深拷贝

m = int(input())

start = []  # 初始图案
for i in range(m):
    a = list(input())
    start.append(a)
start0 = copy.deepcopy(start)
# print('start=',start)

end = []  # 结果图案
for i in range(m):
    a = list(input())
    end.append(a)
# print('end=',end)

ans = [[0 for i in range(m)] for i in range(m)]  # 比较图案


# print('ans=',ans)

def compare(a, b):  # 比较两个图案是否一致
    for i in range(m):
        for j in range(m):
            if a[i][j] != b[i][j]:
                return 0
    return 1


def work1():  # 顺时针转90°
    for i in range(m):
        for j in range(m):
            ans[j][m - i - 1] = start[i][j]
    x = compare(end, ans)
    return x


def work2():  # 顺时针转180°
    for i in range(m):
        for j in range(m):
            ans[m - i - 1][m - j - 1] = start[i][j]
    x = compare(end, ans)
    return x


def work3():  # 顺时针转270°
    for i in range(m):
        for j in range(m):
            ans[m - j - 1][i] = start[i][j]
    x = compare(end, ans)
    return x


def work4():  # 反射(沿着中心轴对调)
    for i in range(m):
        for j in range(m):
            ans[i][m - j - 1] = start[i][j]
    x = compare(end, ans)
    return x


work4()
w4 = copy.deepcopy(ans)


def work5():  # 先对调,然后在方法123中选一个
    for i in range(m):
        for j in range(m):
            ans[j][m - i - 1] = w4[i][j]
    x = compare(end, ans)
    for i in range(m):
        for j in range(m):
            ans[m - i - 1][m - j - 1] = w4[i][j]
    x += compare(end, ans)
    for i in range(m):
        for j in range(m):
            ans[m - j - 1][i] = w4[i][j]
    x += compare(end, ans)
    if x > 0:
        x = 1
    return x


def work6():
    x = compare(start0, end)
    return x


def work():
    if work1():
        print('1')
        return
    if work2():
        print('2')
        return
    if work3():
        print('3')
        return
    if work4():
        print('4')
        return
    if work5():
        print('5')
        return
    if work6():
        print('6')
        return
    print('7')


work()
```
